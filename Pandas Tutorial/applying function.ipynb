```python
import pandas as pd
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['sex_num'] = train.Sex.map({'female' : 0 , 'male' : 1})
```

```python
train.head()
```

```python
train['Name_len'] = train.Name.apply(len)
```

```python
train.head()
```

```python
import numpy as np
train['fare_ceil'] = train.Fare.apply(np.ceil)
```

```python
train.head()
```

```python
train.Name.str.split(',').head()
```

```python
def get_element(my_list , element_no):
    return my_list[element_no]
```

```python
train.Name.str.split(',').apply(get_element,element_no = 0).head()
```

```python
train.Name.str.split(',').apply(lambda x : x[0]).head()
```
