```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def normalise(X):
    mean = np.mean(X)
    std = np.std(X)
    X_copy = (X - mean)/std
    return X_copy
```

```python
data = np.genfromtxt('data.txt' , delimiter = ',')
```

```python
plt.scatter(data[:,0],data[:,1])
plt.show()
```

```python
X_train = data[:60,0]
Y_train = data[:60,1]
X_test = data[60:,0]
Y_test = data[60:,1]
```

```python
plt.scatter(X_train,Y_train , label = 'Training Data' , color = 'r' ,alpha = 0.7)
plt.scatter(X_test,Y_test , label = 'Test Data' , color = 'g' , alpha = 0.7)
plt.legend()
plt.show()
```

```python
X_train.shape
```

```python
X_train_nor = normalise(X_train)
X_test_nor = normalise(X_test)
```

```python
LR = LinearRegression()
LR.fit(X_train_nor.reshape(-1,1) ,Y_train)
```

```python
prediction = LR.predict(X_test_nor.reshape(-1,1))
```

```python
plt.plot(X_test, prediction , label = 'Training Data' , color = 'r')
plt.scatter(X_test,Y_test , label = 'Test Data' , color = 'g' , alpha = .7)
plt.legend()
plt.show()
```
