```python
import requests
r = requests.get('https://www.indeed.co.in/jobs?q=data+scientist&l=')
```

```python
# print the first 500 characters of the HTML
print(r.text[0:500])
```

```python
from bs4 import BeautifulSoup
bs_tree = BeautifulSoup(r.text, 'html.parser')
```

```python
job_link_area = bs_tree.find(id = 'resultsCol')
```

```python
job_link_area
```

```python
job_links = job_link_area.find_all('div')
```
